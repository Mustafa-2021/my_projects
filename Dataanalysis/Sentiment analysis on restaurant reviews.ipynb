```python
import pandas as pd
import numpy as np
import re
import transformers
from transformers import pipeline
```

```python
data = pd.read_csv("Restaurant_reviews.csv") # the data should be in the same folder as your notebook
data.head()
```

```python
data['Review'][0]
```

```python
data.info()
```

```python
#Clean data
# lowercase
data['reviewText_clean'] = data['Review'].str.lower()
data.head()
```

```python
# remove punctuation
#data['reviewText_clean'] = data.apply(lambda x: re.sub(r"([^\w\s])", "", x['reviewText_clean']), axis=1)
data['reviewText_clean'] = data.apply(lambda x: re.sub(r"([^\w\s])", "", str(x['reviewText_clean'])), axis=1)

data.head()
```

```python
sentiment_pipeline = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
```

```python
all_sentiments = []
for review in data['reviewText_clean']:
    result = sentiment_pipeline(review)[0]['label'].lower()
    all_sentiments.append(result)

print(all_sentiments)
```

```python
num_positive = 0
num_negative = 0
for sentiment in all_sentiments:
    if sentiment == 'positive':
        num_positive += 1
    elif sentiment == 'negative':
        num_negative += 1
print(f"There are {num_positive} positive and {num_negative} negative reviews.")
```
